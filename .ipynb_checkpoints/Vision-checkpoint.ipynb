## 이미지 출력

```python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

# prompt = "한국의 음식인 매운 라면 이미지 생성해줘."
prompt = "대표적인 화이트 데스크테리어를 보여줘. 인물은 포함하지 말고 3D로 생성해줘."


response = client.images.generate(
  model="dall-e-3",
  prompt = prompt,
  size="1024x1024",
  quality="standard",
  n=1,
)

image_url = response.data[0].url

# URL에서 이미지를 가져옴
response = requests.get(image_url)

# 응답이 성공적인지 확인
if response.status_code == 200:
    # 이미지 데이터를 바이트로 읽어옴
    image_data = Image.open(BytesIO(response.content))

    # 이미지 표시
    plt.imshow(image_data)
    plt.axis('off')  # 축을 표시하지 않도록 설정
    plt.show()
else:
    print("이미지를 가져오는데 문제가 발생했습니다.")
```

# prompt 영어로 번역 + 이미지 출력

```python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
import tool

# prompt = "한국의 음식인 매운 라면 이미지 생성해줘."
# prompt = "대표적인 화이트 데스크테리어를 보여줘. 여기서 데스크테리어는 데스크와 인테리어의 합성어야 인물은 포함하지 말고 3D로 생성해줘."
prompt = "2100년의 우주정거장의 모습 생성해줘"

# 번역 kor -> eng
#---------------------------------------------------------------------------------
sentence = prompt
sentence = tool.remove_empty_lines(sentence)
# print(sentence)
prompt = f'아래 문장을 영어로 번역해줘.\n\n{sentence}'
print('->prompt: ' + prompt)
  
format = '''
{
  "res": "번역된 문장"
}
'''
  
response = tool.answer('너는 번역기야', prompt, format)
print(response['res']) # {'res': 'Hello. Good morning.'}
# Show me a typical white desk. Create it in 3D without including any characters.
#---------------------------------------------------------------------------------

response = client.images.generate(
  model="dall-e-3",
  prompt = prompt,
  size="1024x1024",
  quality="standard",
  n=1,
)

image_url = response.data[0].url

# URL에서 이미지를 가져옴
response = requests.get(image_url)

# 응답이 성공적인지 확인
if response.status_code == 200:
    # 이미지 데이터를 바이트로 읽어옴
    image_data = Image.open(BytesIO(response.content))

    # 이미지 표시
    plt.imshow(image_data)
    plt.axis('off')  # 축을 표시하지 않도록 설정
    plt.show()
else:
    print("이미지를 가져오는데 문제가 발생했습니다.")
```

## 이미지 저장

```python
# 년월일시분초 + 난수로 파일명 저장
import os
import requests
import random

from datetime import datetime
from PIL import Image
from io import BytesIO

import matplotlib.pyplot as plt

# prompt = "화이트 기반으로 컴퓨터 책상을 꾸민 책상을 보여줘. LED로 방을 좀 더 트렌드하게 해줘. 인물은 포함하지 말고 3D로 생성해줘."
prompt = "화이트 기반으로 컴퓨터 책상 이미지만 3D로 생성해줘, 모니터가 2개이고 벽은 원목으로 생성해줘, 아이스 아메리카노 커피도 책상 위에 추가해줘."

# 번역 kor -> eng
#---------------------------------------------------------------------------------
sentence = prompt
sentence = tool.remove_empty_lines(sentence)
# print(sentence)
prompt = f'아래 문장을 영어로 번역해줘.\n\n{sentence}'
print('->prompt: ' + prompt)
  
format = '''
{
  "res": "번역된 문장"
}
'''
  
response = tool.answer('너는 번역기야', prompt, format)
print(response['res']) # {'res': 'Hello. Good morning.'}
# Show me a typical white desk. Create it in 3D without including any characters.
#---------------------------------------------------------------------------------

# 이미지 생성
#---------------------------------------------------------------------------------
response = client.images.generate(
  model="dall-e-3",
  prompt = prompt,
  size="1024x1024",
  quality="standard",
  n=1,
)

image_url = response.data[0].url

# URL에서 이미지를 가져옴
response = requests.get(image_url)

# 응답이 성공적인지 확인
if response.status_code == 200:
    # 이미지 데이터를 바이트로 읽어옴
    image_data = Image.open(BytesIO(response.content))

    # 이미지 표시
    plt.imshow(image_data)
    plt.axis('off')  # 축을 표시하지 않도록 설정
    plt.show() # 이미지 출력
else:
    print("이미지를 가져오는데 문제가 발생했습니다.")
#---------------------------------------------------------------------------------

# 이미지 저장
#---------------------------------------------------------------------------------
# 현재 시간을 가져옴
now = datetime.now()

# '년월일시분초' 형식의 문자열 생성
date_time_string = now.strftime("%Y%m%d%H%M%S")

# 1부터 1000까지의 난수 생성
random_number = random.randint(1, 1000)

if(os.path.exists('./static/storage') == False):
    os.mkdir('./static/storage')
    
# 파일명 생성 (년월일시분초_난수.txt 형식)
file_name = f"./static/storage/{date_time_string}_{random_number}.jpg"

# 응답이 성공적인지 확인
if response.status_code == 200:
    # 이미지 데이터를 파일로 저장
    with open(file_name, "wb") as file:
        file.write(response.content)
    print("이미지가 성공적으로 저장되었습니다.")
else:
    print("이미지를 가져오는데 문제가 발생했습니다.")

#---------------------------------------------------------------------------------
```

## 이미지 인식

```python
from openai import OpenAI

client = OpenAI()

prompt = "이미지에 무엇이 있는지 알려줘"
url = "https://upload.wikimedia.org/wikipedia/commons/thumb/d/dd/Gfp-wisconsin-madison-the-nature-boardwalk.jpg/2560px-Gfp-wisconsin-madison-the-nature-boardwalk.jpg"
response = client.chat.completions.create(
  model="gpt-4o",
  messages=[
    {
      "role": "user",
      "content": [
        {"type": "text", "text": prompt},
        {
          "type": "image_url",
          "image_url": {
            "url": url,
          },
        },
      ],
    }
  ],
  max_tokens=300,
)

print(response.choices[0].message.content)
```

# 이미지 업로드 및 이미지 인식

```python
import os
import base64
import requests

# OpenAI API Key
api_key = os.getenv('OPENAI_API_KEY')


# Function to encode the image
def encode_image(image_path):
  with open(image_path, "rb") as image_file:
    return base64.b64encode(image_file.read()).decode('utf-8')

# Path to your image
image_path = "./static/storage/food/budae01.jpg" # 부대찌개

# Getting the base64 string
base64_image = encode_image(image_path)

headers = {
  "Content-Type": "application/json",
  "Authorization": f"Bearer {api_key}"
}

prompt  = "메뉴 알려줘"
payload = {
  "model": "gpt-4o",
  "messages": [
    {
      "role": "user",
      "content": [
        {
          "type": "text",
          "text": prompt
        },
        {
          "type": "image_url",
          "image_url": {
            "url": f"data:image/jpeg;base64,{base64_image}"
          }
        }
      ]
    }
  ],
  "max_tokens": 300
}

response = requests.post("https://api.openai.com/v1/chat/completions", headers=headers, json=payload)

print(response.json()['choices'][0]['message']['content'])
```

```python
import os
import base64
import requests
import json

# OpenAI API Key
api_key = os.getenv('OPENAI_API_KEY')


# Function to encode the image
def encode_image(image_path):
  with open(image_path, "rb") as image_file:
    return base64.b64encode(image_file.read()).decode('utf-8')

# Path to your image
image_path = "./static/storage/food/food17.jpg" # 삼겹살

# Getting the base64 string
base64_image = encode_image(image_path)

headers = {
  "Content-Type": "application/json",
  "Authorization": f"Bearer {api_key}"
}

prompt  = "메뉴 알려줘"
payload = {
  "model": "gpt-4o",
  "messages": [
    {
      "role": "user",
      "content": [
        {
          "type": "text",
          "text": prompt + '\n\n출력 형식: {"메뉴명": "메뉴 이름"}'
        },
        {
          "type": "image_url",
          "image_url": {
            "url": f"data:image/jpeg;base64,{base64_image}"
          }
        }
      ]
    }
  ],
  "max_tokens": 300

}

response = requests.post("https://api.openai.com/v1/chat/completions", headers=headers, json=payload)

print(response.json())
print(response.json()['choices'][0]['message']['content'])
menu = response.json()['choices'][0]['message']['content'] # 메뉴명 추출
print(json.loads(menu)['메뉴명'])

prompt = menu + '맛있게 만드는 방법 알려줘'
```
